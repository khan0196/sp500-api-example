```python
import yfinance as yf
import datetime as dt
import pandas as pd
import numpy as np

```

```python
# scrap the current member of S&P500 from wikipedia
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
data = pd.read_html(url)
```

```python
sp500  = data[0].iloc[:,[0,1,6,7]]
sp500.columns = ['ticker','name','date','cik']
sp500 = sp500[~sp500['ticker'].isnull()]
sp500.head()
```

sp500_sorted = sp500.copy()
sp500_sorted = sp500.iloc[ sp500_sorted.sort_values(by=['ticker'],ascending =True).index.values,:].copy()
sp500_sorted

```python
sp500_sorted = sp500.copy()
sp500_sorted = sp500_sorted.sort_values(by=['ticker'],ascending =True,ignore_index = True)
sp500_sorted.head()
```

```python
test = sp500_sorted['ticker'].tolist()
```

```python
# retrieve data from yahoo finance
stocks = yf.download(test, start="2000-01-01", end="2021-09-01")
stocks
```

```python
# take adjusted closing price only
stocks_adj = stocks.copy()
stocks_adj = stocks_adj.iloc[:,0:505]
stocks_adj.head()
```

```python
stocks_adj[stocks_adj.iloc[:,0].isnull()].index
```

```python
stocks_adj = stocks_adj.drop(axis=0, labels =stocks_adj[stocks_adj.iloc[:,0].isnull()].index)
```

```python
stocks_adj[stocks_adj.iloc[:,3].isnull()].index
```

```python
# 363 members in S&P 500 in the market since 2000
l = stocks_adj.isnull().sum()==0
sum(l)
```

```python
# remove stocks not in the market since 2000
stocks_adj_final = stocks_adj.copy()
stocks_adj_final = stocks_adj_final.iloc[:,l.tolist()]
stocks_adj_final.head()
```

```python
sum(stocks_adj_final.isnull().sum())
```

```python
c= stocks_adj_final.columns.copy()
c = c.to_frame()
c = c.loc[:,1].tolist()
```

```python
stocks_adj_final.columns = c
stocks_adj_final.head()
```

```python
stocks_logrt = stocks_adj_final.copy()
stocks_logrt = np.log(stocks_logrt) - np.log(stocks_logrt.shift(1))
stocks_logrt = stocks_logrt.drop(pd.Timestamp('2000-01-03 00:00:00'))
```

```python
stocks_logrt.to_csv('sp500.csv',index=True)
```
